### Unwanted Class Folder Deletion

```python
import os
import shutil

base_path = "raw_data"

unwanted_folders = ["receipts", "qr", "symbols"]

for folder in unwanted_folders:
    folder_path = os.path.join(base_path, folder)
    if os.path.exists(folder_path):
        shutil.rmtree(folder_path)
        print(f"Deleted folder: {folder_path}")
    else:
        print(f"Not found: {folder_path}")
```

### corrupted and unreadable images

```python
from PIL import Image
import os

def remove_corrupted_images(folder):
    for root, dirs, files in os.walk(folder):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                img = Image.open(file_path)
                img.verify()  # check if valid
            except:
                print(f"Removing corrupted: {file_path}")
                os.remove(file_path)

remove_corrupted_images("raw_data")
```

### Remove duplicate images

```python
import os
import hashlib

def get_hash(file_path):
    with open(file_path, 'rb') as f:
        return hashlib.md5(f.read()).hexdigest()

def remove_duplicates(folder):
    hashes = {}
    for root, dirs, files in os.walk(folder):
        for file in files:
            file_path = os.path.join(root, file)
            file_hash = get_hash(file_path)
            
            if file_hash in hashes:
                print(f"Duplicate found, deleting: {file_path}")
                os.remove(file_path)
            else:
                hashes[file_hash] = file_path

remove_duplicates("raw_data")
```

```python
import os
import shutil

source = "raw_data"
dest_real = "cleaned_data/real"
dest_fake = "cleaned_data/fake"

os.makedirs(dest_real, exist_ok=True)
os.makedirs(dest_fake, exist_ok=True)

real_keywords = ["real", "true", "original", "genuine"]
fake_keywords = ["fake", "forged", "tampered"]

count_real = 0
count_fake = 0

for root, dirs, files in os.walk(source):
    for file in files:
        if file.lower().endswith(('.jpg', '.png', '.jpeg')):
            path = os.path.join(root, file)
            folder_name = root.lower()

            # Check for fake
            if any(word in folder_name for word in fake_keywords):
                shutil.copy(path, dest_fake)
                count_fake += 1

            # Check for real
            elif any(word in folder_name for word in real_keywords):
                shutil.copy(path, dest_real)
                count_real += 1

print("Copied Real:", count_real)
print("Copied Fake:", count_fake)
```
